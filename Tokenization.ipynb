```python
import nltk
```

```python
nltk.download()
```

```python
text = "Hello! How are you donig today? I hope you're having a good day."
```

## Tokenization the text into individual sentences

```python
sentences = nltk.sent_tokenize(text)
```

```python
tokens = nltk.word_tokenize(text)
```

```python
print(sentences)
```

```python
print(tokens)
```
